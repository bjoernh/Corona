```python
dict_pavel = pavel.set_index("Landkreis")["Kontaktrisiko"]
dict_pavel['SK Augsburg']
del(dict_pavel)
```

```python
m = folium.Map(location=[51.39699005966058, 10.283880160916578], prefer_canvas=True, zoom_start=6, tiles='CartoDB positron')
m2 = folium.Map(location=[51.39699005966058, 10.283880160916578], zoom_start=6, tiles='CartoDB positron')
```

```python
def style_pavel(feature):
        c = colormap(feature['properties']['Kehrwert_risiko'])
        return {"fillColor": c,
        "color": "white",
        "weight": 1,
        "dashArray": "5, 5",
        "fillOpacity": 0.8,}
```

```python
colormap2 = linear.Reds_05.scale(
    rki.cases7_per_100k.min(), rki.cases7_per_100k.max()
)
def style_rki(feature):
        c = colormap2( int(feature["properties"]["cases7_per_100k"]))
        return {"fillColor": c,
        "color": "white",
        "weight": 1,
        "dashArray": "5, 5",
        "fillOpacity": 0.8,}
```

```python
tooltip = GeoJsonTooltip(
    fields=["Landkreis", "InzidenzFallNeu-7-Tage", "Kontaktrisiko", "MeldeDauerFallNeu-Schnitt"],
    aliases=["Landkreis:",  "Inzidenz:", "Kontaktrisiko:", "Ø Meldedauer:"],
    localize=True,
    sticky=False,
    labels=True,
    style="""
        background-color: #F0EFEF;
        border: 2px solid black;
        border-radius: 3px;
        box-shadow: 3px;
    """,
    max_width=800,
)
```

```python
folium.GeoJson(
    sub[sub.geometry != None],
    name="Pavel",
    style_function=style_pavel,
    tooltip=tooltip,
).add_to(m)

folium.LayerControl().add_to(m)
#colormap.caption = "Kontaktrisiko"
#colormap.add_to(m)
m
```

```python
folium.GeoJson(
    rki,
    name="rki",
    style_function=style_rki,
).add_to(m2)

folium.LayerControl().add_to(m2)
```
